### Imports + Load Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
# Load UNICEF dataset
df_trend = pd.read_csv("../data/processed/india_u5mr_full.csv")

df_trend.head()
```

```python
df_trend.tail()
```

```python
print("Shape:", df_trend.shape)
df_trend.info()
df_trend.describe()
```

```python
df_trend.isnull().sum()
```

### U5MR Over Time

```python
plt.figure(figsize=(10,6))

sns.lineplot(
    data=df_trend,
    x='year',
    y='u5mr'
)

plt.title("Under-5 Mortality Rate (India) Over Time")
plt.xlabel("Year")
plt.ylabel("U5MR (per 1000 live births)")

plt.savefig("../outputs/figures/trends/unicef_u5mr_trend_v1.png",
            dpi=300,
            bbox_inches='tight')

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.lineplot(
    data=df_trend,
    x='year',
    y='u5mr',
    marker='o'
)

plt.title("Trend of Under-5 Mortality Rate in India")
plt.xlabel("Year")
plt.ylabel("U5MR")

plt.savefig("../outputs/figures/trends/unicef_u5mr_trend_marked_v1.png",
            dpi=300,
            bbox_inches='tight')

plt.show()
```

```python
start_value = df_trend.iloc[0]['u5mr']
end_value = df_trend.iloc[-1]['u5mr']

reduction = ((start_value - end_value) / start_value) * 100

print("Start U5MR:", start_value)
print("End U5MR:", end_value)
print("Percentage Reduction:", reduction)
```

### Focus on recent years (2000 onwards)

```python
df_recent = df_trend[df_trend['year'] >= 2000]

df_recent.head()
```

```python
plt.figure(figsize=(10,6))

sns.lineplot(
    data=df_recent,
    x='year',
    y='u5mr',
    marker='o'
)

plt.title("Under-5 Mortality Rate (India) Since 2000")
plt.xlabel("Year")
plt.ylabel("U5MR")

plt.savefig("../outputs/figures/trends/unicef_recent_trend_v1.png",
            dpi=300,
            bbox_inches='tight')

plt.show()
```
