### Imports + Load Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
# Load datasets
df_causes = pd.read_csv("../data/processed/child_death_causes_srs.csv")
df_category = pd.read_csv("../data/processed/child_death_category_srs.csv")

df_causes.head(), df_category.head()
```

```python
print("Causes shape:", df_causes.shape)
print("Category shape:", df_category.shape)

df_causes.info()
df_category.info()
```

```python
df_causes.isnull().sum(), df_category.isnull().sum()
```

### Detailed Causes Visualization

```python
plt.figure(figsize=(10,6))

sns.barplot(
    data=df_causes,
    x='percentage',
    y='cause'
)

plt.title("Major Causes of Child Death (SRS Data)")
plt.xlabel("Percentage")
plt.ylabel("Cause")

plt.savefig("../outputs/figures/comparisons/srs_causes_distribution_v1.png",
            dpi=300,
            bbox_inches='tight')

plt.show()
```

### Category-Level Visualization

```python
plt.figure(figsize=(8,5))

sns.barplot(
    data=df_category,
    x='percentage',
    y='category'
)

plt.title("Child Death Causes by Category (SRS)")
plt.xlabel("Percentage")
plt.ylabel("Category")

plt.savefig("../outputs/figures/comparisons/srs_category_bar_v1.png",
            dpi=300,
            bbox_inches='tight')

plt.show()
```

### Pie Chart

```python
plt.figure(figsize=(6,6))

plt.pie(
    df_category['percentage'],
    labels=df_category['category'],
    autopct='%1.1f%%'
)

plt.title("Distribution of Child Death Causes")

plt.savefig("../outputs/figures/comparisons/srs_category_pie_v1.png",
            dpi=300,
            bbox_inches='tight')

plt.show()
```
